Contiguous list
Description
Given a list as input, your task is to check if the list is strictly increasing. i.e. the numbers in the list should be in an increasing order. Hence, a number at a lower index should always be smaller than a number at a higher index.
Assume that the list will only have positive integers.
Print "yes" if the list is in strictly increasing order and print "no" if the list is not strictly increasing. 

Sample Input:
[1,4,300,400,900]
Sample Output:
yes

This question was asked in a Data Analyst interview.

```python
import ast,sys
input_str = sys.stdin.read()
input_list = ast.literal_eval(input_str)
length= len(input_list)
li= []

for i in range(0,length-1):
    if (input_list[i+1]-input_list[i])>0:
        li.append('yes')
    else:
        li.append('no')
        
            
#print(li)
if 'no' in li:
    print('no')
else:
    print('yes')
        
```

Vote for Food
Description
Your team is going for camping and you are taking a vote to decide what food to pack for dinner.
Everyone gets a vote and the food item that gets at least one more than half of the votes wins. None of the items wins if nothing gets at least one more than half votes. Assume that every person gets only one vote.
The input will contain a list of food items where each occurrence of an item represents one vote. You should print the winning food item as output. If there is no clear winner, print "NOTA". 

Sample Input:
["pasta","pasta","pasta","pasta","pasta","paratha","paratha","paratha"]
Sample Output:
pasta

This question was asked in a Data Analyst interview.

```python
import ast,sys
input_str = input()
votes = ast.literal_eval(input_str)

votes1 = votes

def most_common(lst): 
    return max(set(lst), key=lst.count) 

m = most_common(votes1) 


wordfreq = [] 

for w in votes1: 
    wordfreq.append(votes1.count(w)) 
    
print(wordfreq)
    
if max(wordfreq) > len(wordfreq)/2:
    print(m) 
else: 
    print("NOTA")
```

Knapsack
Description
You were a businessman, but you recently suffered a big loss and are now bankrupt. You turn to the dark side and become a thief. You enter a house and find that there are ‘n’ items worth stealing in the house, but unfortunately, your bag is too small to carry ALL the items.

 

You now have to now select the items such that you get the maximum value. Each item ‘i’ is of value ‘vi’ and weight ‘wi’. But since your bag capacity(weight) is ‘B’, you can only take items such that the sum of the spaces of items you decide to take is less than or equal to B. You can break the items to maximize the total value of Knapsack. 



Input Format:

The first element of the list will be a positive integer, bag size B.
 The second element of the list will be a list of length ‘n’ of positive integers, denoting the weight each item takes.
 The third element of the list will be the respective values of the ‘n’ items in the same order of items as the previous list.
Output Format:

Maximum value of the knapsack in round figure.





Examples:

Input 1:

[60,[11, 22, 30],[65, 45, 100] ]



Output 1:

204



Explanation: The maximum capacity of Knapsack is 60, the weights of the items to be stolen are 11,22,30 and its respective values equals to 65, 45, 100. The maximum value in Knapsack can be 204. The knapsack will contain the weight 30 of value 100, weight 11 of value 65 and partial weight of 22 for the remaining value that is equal to 39 (= 19/22 * 45). Thus it adds to the value that is equal to 204.

```python
import ast,sys
import math
input_str = sys.stdin.read()
input_list = ast.literal_eval(input_str)

B=input_list[0] #Bag size
w=input_list[1] #list of weight of each item
val=input_list[2] #list of values of each item 

#type your code here
final_v = 0
b=B
while(b > 0):
    max_val = max(val)
    indx = val.index(max_val)
    weight = w[indx]
    if(b > weight):
        b = b-weight
        final_v = final_v+ max_val
        val.pop(indx)
        w.pop(indx)
    else:
        per_unit = val[indx]/w[indx]
        temp = b*per_unit
        final_v = final_v+temp
        b = b-weight

print(math.ceil(final_v))
```

```python
Val = {range(10):range(48, 58), range(10,36): range(65, 91)}

def B2D(N,B1):
    '''From Base B1 to Decimal'''
    DN = 0
    FV=0
    for i in range(len(N)):
        for j in Val:
            if ord(N[i]) in Val[j]:
                FV=j[ord(N[i])-Val[j][0]]       # FaceValue of the Digit
        if FV>= B1:                             # Digits aren't >=Base, right?
            print("Base Error..")
            exit()
        else:
            DN += FV * (B1 ** (len(N) - 1 - i))
    return DN

def D2B(DN,B2):
    '''From Decimal to Base B2'''
    if int(DN) == 0:
        BN = '0'
    else:
        BN = ""
        while DN > 0:
            R = DN % B2
            for i in Val:
                if R in i:
                    BN+=chr(Val[i][R-i[0]])     #Finding the Digit for the Value
            DN = int(DN / B2)
    return BN[::-1]

def B2B(N,B1,B2):
    return D2B(B2D(N,B1),B2)

N=input("Num:")
B1=int(input("FromBase:"))
B2=int(input("ToBase:"))
print("Base[",B1,"]:",N)
print("Decimal:",B2D(N,B1))
print("Base[",B2,"]:",B2B(N,B1,B2))
```
